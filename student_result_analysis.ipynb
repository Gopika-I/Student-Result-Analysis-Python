Gender Distribution

```python
plt.figure(figsize=(6,6))
sns.countplot(data=data, x="Gender")
plt.title("Gender Distribution")
plt.show()
```

Grouping the data based on Parents Education

```python
Parent_education = data.groupby("ParentEduc").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
Parent_education
```

```python
plt.figure(figsize=(6,6))
sns.heatmap(Parent_education, annot=True)
plt.title("Relationship between parents education and students score")
plt.show()
```

There is relationship between Parents Education and Students score.

```python
ParentMaritalStatus = data.groupby("ParentMaritalStatus").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
ParentMaritalStatus
```

```python
plt.figure(figsize=(6,6))
sns.heatmap(ParentMaritalStatus, annot=True)
plt.title("Relationship between Parents Marital Status and students score")
plt.show()
```

There is no significant relationship between Parents marital Status and Students score.

```python
WklyStudyHours=data.groupby("WklyStudyHours").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
WklyStudyHours
```

```python
plt.figure(figsize=(4,4))
sns.heatmap(WklyStudyHours,annot=True)
plt.title("Relationship between Weekly Study Hours and Student's Scores")
plt.show()
```

There is a relationship between wkly Study hours and Students score.

```python
PracticeSport= data.groupby("PracticeSport").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
PracticeSport
```

```python
plt.figure(figsize=(4,4))
sns.heatmap(PracticeSport,annot=True)
plt.title("Relationship between Sports Practicing and Student's Score")
plt.show()
```

There is no relationship between Sports Practicing and Student's Score

```python
groupA= data.loc[(data['EthnicGroup'] == "group A")].count()
groupB= data.loc[(data['EthnicGroup'] == "group B")].count()
groupC= data.loc[(data['EthnicGroup'] == "group C")].count()
groupD= data.loc[(data['EthnicGroup'] == "group D")].count()
groupE= data.loc[(data['EthnicGroup'] == "group E")].count()

mlist = [groupA["EthnicGroup"], groupB["EthnicGroup"], groupC["EthnicGroup"], groupD["EthnicGroup"], groupE["EthnicGroup"]]
l = ["groupA","groupB","groupC","groupD","groupE"]
plt.pie(mlist,labels = l,autopct = "%1.2f%%")
plt.title("Distribution of Ethnic Groups")
plt.show()
```
